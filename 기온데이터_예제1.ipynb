```python
# s.csv 파일의 데이터 구조를 확인할 것

import csv
f = open('s.csv', encoding='cp949') 
data = csv.reader(f)

next(data)   

result = []  # 최고 기온 데이터를 저장할 리스트 생성

for row in data :
    if row[-1] != '' :                # 최고 기온 데이터 값이 존재한다면
        result.append(float(row[-1])) # result 리스트에 최고 기온 값 추가
print(result)

# 그래프
import matplotlib.pyplot as plt

plt.figure(figsize = (10,3), dpi = 300)
plt.plot(result, 'r')
plt.show()
```

```python
# 원하는 날짜 데이터 추출하기

import csv

f = open('s.csv', encoding='cp949') 
data = csv.reader(f)
next(data)
result = []

for row in data :
    if row[-1] != '' :
        if row[0].split('-')[1] == '08' :   # 8월 데이터 만 추출
            result.append(float(row[-1]))

# 시각화

import matplotlib.pyplot as plt

plt.figure(dpi = 100)
plt.plot(result, 'hotpink')
plt.show()
```

```python
 # 1월1일과 9월15일 날씨 비교

import csv

f = open('s.csv', encoding='cp949')
data = csv.reader(f)
next(data)
new_year = []
thanks_day = []

for row in data :
    if row[-1] != '' and row[-2] != '' : # 값이 존재하면 
        if 1990 <= int(row[0].split('-')[0]) : 
          if row[0].split('-')[1] == '01' and row[0].split('-')[2] == '01' :
                new_year.append(float(row[-1]))
          if row[0].split('-')[1] == '09' and row[0].split('-')[2] == '15' :      
                thanks_day.append(float(row[-1]))


import matplotlib.pyplot as plt

plt.rc('font', family='NanumGothic') 
plt.figure(figsize=(10,5))
plt.title('명절 기온변화 그래프', size = 20)

#plt.figure(figsize=(10,5) )
plt.plot(new_year, 'hotpink', label='설날')
plt.plot(thanks_day, 'skyblue', label='한가위')

plt.legend( loc= 7)
plt.show()
```
