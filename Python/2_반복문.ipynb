```
#중첩 for 문 # %2d (자리수 2개를 고정값으로-줄을 맞춰주기 위해서)
## 전역 변수 선언 부분 ##
i, k, guguLine = 0, 0, ""

## 메인 코드 부분 ##
for i in range(2, 10) :
     guguLine = guguLine + (" ##  %d단  ##" % i)

print(guguLine)

for i in range(1, 10) :
     guguLine = ""
     for k in range(2, 10) :
          guguLine = guguLine + str("%2d X %2d = %2d\t" % (k, i, k*i))
     print(guguLine)
```

```
#별 찍기 문제
#i 가 j보다 크거나 같을 때, 별을 찍는다.

print ("1번문제")
for i in range(0, 5) : 
    #(0,5) = 0,1,2,3,4 출력
    for j in range (0,5):
      if i>=j :
        print("%3c"%"*", end="")
      else :
        print("%3c" % " ", end="")
    print()

print("---------------------")

print ("2번문제")
for i in range(0, 5) : 
    #(0,5) = 0,1,2,3,4 출력
    for j in range (0,5):
      if j>=i :
        print("%3c"%"*", end="")
      else :
        print("%3c" % " ", end="")
    print()


print("---------------------")

print ("3번문제")
for i in range(0, 5) : 
    #(0,5) = 0,1,2,3,4 출력
    for j in range (0,5):
      if i+j<=4 :
        print("%3c"%"*", end="")
      else :
        print("%3c" % " ", end="")
    print()

print("---------------------")

print ("4번문제")
for i in range(0, 5) : 
    #(0,5) = 0,1,2,3,4 출력
    for j in range (0,5):
      if j+i>=4 :
        print("%3c"%"*", end="")
      else :
        print("%3c" % " ", end="")
    print()

```

```
#1~10 까지의 합

i, hap=0,0

i=1
while i<11:
  hap=hap+i
  i=i+1

print("1에서 10까지의 합계:%d" % hap)
```
